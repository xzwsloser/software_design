```python
from typing import List

# 词嵌入
def word2vec(text: List[str]) -> List[np.array]:
    resp = dashscope.TextEmbedding.call(
        model="text-embedding-v4",
        input=text,
        dimension=256
    )

    if not (resp.status_code == HTTPStatus.OK):
        return []

    result = [embed['embedding'] for embed in resp.output['embeddings']]
    return result
```

```python
from tqdm import tqdm

embed_introduce_list = []
for idx, cur_text in tqdm(enumerate(introduce_text_list)):
    embed_vec = word2vec([cur_text])
    print('='*10 + f'{idx + 1}' + '='*10)
    embed_introduce_list.append(embed_vec[0])
```
