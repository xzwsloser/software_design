# 评论数据分析

```python
import pandas as pd
import numpy as np
```

```python
positive_comment_infos = pd.read_csv('../dataset/tb_positive_comment_extra2.csv')
```

```python
positive_comment_infos.head()
```

```python
ip_columns = positive_comment_infos['ip']
ip_columns.head()
```

```python
ip_counts = ip_columns.value_counts().sort_index()
ip_counts, len(ip_counts)
```

```python
tourist_type_columns = positive_comment_infos['tourist_type']
tourist_type_columns.head()
```

```python
tourist_type_counts = tourist_type_columns.value_counts().sort_index()
tourist_type_counts, len(tourist_type_counts)
```

```python
len(tourist_type_columns[tourist_type_columns.isna()].index)
```

```python
content_columns = positive_comment_infos['content']
content_columns.head()
```

```python
content_len_columns = content_columns.map(lambda x: len(x) if isinstance(x, str) else 0)
content_len_columns.head()
```

```python
content_len_columns.describe()
```

```python
negative_comment_infos = pd.read_csv('../dataset/tb_negative_comment_extra.csv')
negative_comment_infos.head()
```

```python
negative_content_columns = negative_comment_infos['content']
negative_content_columns.head()
```

```python
negative_content_len_columns = negative_content_columns.map(lambda x: len(x) if isinstance(x, str) else 0)
negative_content_len_columns.head()
```

```python
negative_content_len_columns.describe()
```
